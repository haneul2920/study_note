```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_pepTestCustomers.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
age_avg = round(df.age.mean(),0)
df['age'] = df['age'].fillna(age_avg)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
import sklearn

# 1. X 값들 과 Y값을 추출한다.
# 2. train_test_split(X값들, Y값, test_size=%값)
# 3. 결과 : 학습용 x_train, x_test, y_train, y_test
```

```python
# 1. X 값들 과 Y값을 추출한다.
# df.head(3)
dfx = df.drop(['id', 'pep'], axis=1)
dfy = df['pep']
```

```python
x_train, x_test, y_train, y_test = train_test_split(dfx, dfy, test_size=0.25, random_state=0)
```

```python
print(df.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Modeling

```python
# 예측 알고리즘
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=6, random_state=0)
```

```python
tree.fit(x_train, y_train)
```

```python
pred_tree = tree.predict(x_test)
pred_tree
```

```python
tree.score(x_test, y_test)
```

### 적용(전개) Deployment

```python
ndf = pd.read_csv('data_pepNewCustomers.csv')
ndf
```

```python
ndf.head(3)
```

```python
ndf.loc[:,'age':'mortgage']
```

```python
ndf['pred'] = tree.predict(ndf.loc[:,'age':'mortgage'])
```

```python
ndf.head(20)
```

```python
# 예측결과들에 대한 확률값
ndf['pred_proba'] = tree.predict_proba(ndf.loc[:,'age':'mortgage'])[:,1]
ndf.head(20)
```

```python
# pred == 1 & pred_proba > 0.7 조건을 만족하는 리스트를 구하자
# pep에 가입할  확률이 70%가 넘는 고객들

target = ndf.query('pred == 1 & pred_proba > 0.7')
target
```

```python
# 확률이 높은 것부터 내림차순으로 정리
# CSV 파일로 추출

target.sort_values(by='pred_proba', ascending=False).to_csv('pep_target_김한울.csv', index=False)
```
