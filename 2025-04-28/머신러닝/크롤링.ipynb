```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Chrome 드라이버 실행
driver = webdriver.Chrome()

# "전체상품" 페이지 열기
base_url = 'https://www.mychef.kr/product/list?depth1=All'
driver.get(base_url)

# 페이지 로딩 대기
time.sleep(3)

page = 1  # 페이지 번호 초기화
clicked_urls = set()  # 이미 클릭한 상품의 URL을 추적할 set

try:
    while True:  # 페이지가 더 이상 없을 때까지 반복
        print(f"\n🚀 {page} 페이지 크롤링 시작 🚀")

        while True:  # 상품 클릭 후 리스트를 다시 불러오기 위해 별도 루프
            try:
                # 현재 페이지의 상품 리스트 새로 가져오기
                product_elements = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
                )
                break  # 성공적으로 상품을 가져오면 반복문 탈출
            except:
                print("❌ 상품 목록을 불러오지 못했습니다. 다시 시도합니다.")
                time.sleep(2)

        print(f"✅ {page} 페이지에서 {len(product_elements)}개의 상품 발견")

        # 상품 URL 리스트 저장 (중복 클릭 방지)
        product_urls = [product.get_attribute("href") for product in product_elements]

        for index, product_url in enumerate(product_urls):
            if product_url in clicked_urls:
                print(f"🔴 이미 클릭한 상품: {product_url} - 건너뜁니다.")
                continue  # 이미 클릭한 상품은 건너뜁니다.

            try:
                print(f"▶ {index + 1}번째 상품 클릭: {product_url}")
                driver.get(product_url)  # 상품 페이지로 직접 이동
                time.sleep(2)  # 상세 페이지 로딩 대기

                clicked_urls.add(product_url)  # 클릭한 URL을 기록

                driver.back()  # 다시 전체상품 페이지로 이동
                time.sleep(2)  # 페이지 로딩 대기

                # 🔥 상품 리스트 새로 가져오기 (중복 클릭 방지)
                while True:
                    try:
                        product_elements = WebDriverWait(driver, 10).until(
                            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
                        )
                        break  # 성공적으로 로딩되면 탈출
                    except:
                        print("🔄 상품 목록을 다시 불러오는 중...")
                        time.sleep(2)

            except Exception as e:
                print(f"❌ 오류 발생: {e}")
                continue  # 다음 상품으로 진행

        # 다음 페이지로 이동 (JavaScript 실행)
        try:
            page += 1  # 다음 페이지
            driver.execute_script(f"cmPageMove('{page}')")  # JavaScript 실행
            time.sleep(3)  # 페이지 로딩 대기
        except:
            print("❌ 더 이상 다음 페이지가 없습니다.")
            break  # 페이지가 없으면 종료

except Exception as e:
    print("❌ 크롤링 중 오류 발생:", e)

# 브라우저 종료
driver.quit()
```

```python
import os
import csv
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 저장할 폴더 생성
image_folder = "product_images"
if not os.path.exists(image_folder):
    os.makedirs(image_folder)

# CSV 파일 설정
csv_filename = "reviews.csv"
csv_columns = ["상품 URL", "상품 이름", "리뷰 ID", "별점", "리뷰 내용"]

# CSV 파일 초기화 (첫 실행 시 헤더 작성)
if not os.path.exists(csv_filename):
    with open(csv_filename, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(csv_columns)

# Chrome 드라이버 실행
driver = webdriver.Chrome()

# "전체상품" 페이지 열기
base_url = 'https://www.mychef.kr/product/list?depth1=All'
driver.get(base_url)

# 페이지 로딩 대기
time.sleep(3)

page = 1  # 페이지 번호 초기화
clicked_urls = set()  # 이미 크롤링한 상품의 URL을 추적할 set

try:
    while True:  # 페이지가 더 이상 없을 때까지 반복
        print(f"\n🚀 {page} 페이지 크롤링 시작 🚀")

        while True:  # 상품 클릭 후 리스트를 다시 불러오기 위해 별도 루프
            try:
                # 현재 페이지의 상품 리스트 새로 가져오기
                product_elements = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
                )
                break  # 성공적으로 상품을 가져오면 반복문 탈출
            except:
                print("❌ 상품 목록을 불러오지 못했습니다. 다시 시도합니다.")
                time.sleep(2)

        print(f"✅ {page} 페이지에서 {len(product_elements)}개의 상품 발견")

        # 상품 URL 리스트 저장 (중복 클릭 방지)
        product_urls = [product.get_attribute("href") for product in product_elements]

        for index, product_url in enumerate(product_urls):
            if product_url in clicked_urls:
                print(f"🔴 이미 크롤링한 상품: {product_url} - 건너뜁니다.")
                continue  # 이미 크롤링한 상품은 건너뜁니다.

            try:
                print(f"▶ {index + 1}번째 상품 크롤링: {product_url}")
                driver.get(product_url)  # 상품 페이지로 직접 이동
                time.sleep(2)  # 상세 페이지 로딩 대기

                # 🔹 1. 상품 이름 가져오기
                try:
                    product_name = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, ".goods-tit"))
                    ).text.strip()
                    print(f"🛒 상품 이름: {product_name}")
                except:
                    product_name = "상품 이름 없음"

                # 🔹 2. 상품 이미지 저장
                try:
                    image_element = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, "img.lozad"))
                    )
                    image_url = image_element.get_attribute("src")

                    if image_url:
                        image_name = os.path.join(image_folder, f"{index + 1}_image.jpg")
                        img_data = requests.get(image_url).content
                        with open(image_name, "wb") as handler:
                            handler.write(img_data)
                        print(f"🖼️ 이미지 저장 완료: {image_name}")
                except Exception as e:
                    print(f"❌ 이미지 저장 실패: {e}")

                # 🔹 3. 리뷰 ID, 내용, 별점 크롤링 후 CSV 저장
                try:
                    review_elements = driver.find_elements(By.CSS_SELECTOR, "[li.name.user-name span]")
                    reviews = []

                    for review_element in review_elements:
                        # 리뷰 ID
                        review_id = review_element.get_attribute("li.name.user-name span")

                        # 리뷰 내용
                        try:
                            review_text = review_element.find_element(By.CSS_SELECTOR, "p.review.text span").text.strip()
                        except:
                            review_text = "내용 없음"
                            print('내용에서 실패')

                        # 리뷰 별점
                        try:
                            star_element = review_element.find_element(By.CSS_SELECTOR, "a.pointReviewContent span")
                            star_style = star_element.get_attribute("style")  # 예: "width:95px"
                            star_match = re.search(r'width:\s*(\d+)px', star_style)

                            if star_match:
                                star_width = int(star_match.group(1))  # 숫자만 추출
                                star_rating = round(star_width / 20, 1)  # 19px = 1점
                            else:
                                star_rating = "별점 없음"
                        except:
                            star_rating = "별점 없음"
                            print('별점에서 실패')

                        reviews.append([product_url, product_name, review_id, star_rating, review_text])

                    if reviews:
                        # CSV 파일에 저장
                        with open(csv_filename, "a", newline="", encoding="utf-8-sig") as f:
                            writer = csv.writer(f)
                            writer.writerows(reviews)

                        for review in reviews:
                            print(f"💬 리뷰 저장: 상품={review[1]} | ID={review[2]} | ⭐ 별점={review[3]} | ✍ 내용={review[4]}")
                    else:
                        print("💬 리뷰 없음")

                except Exception as e:
                    print(f"❌ 리뷰 크롤링 실패: {e}")

                clicked_urls.add(product_url)  # 크롤링 완료한 상품 URL 기록

                driver.back()  # 다시 전체상품 페이지로 이동
                time.sleep(2)  # 페이지 로딩 대기

                # 🔥 상품 리스트 새로 가져오기 (중복 크롤링 방지)
                while True:
                    try:
                        product_elements = WebDriverWait(driver, 10).until(
                            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
                        )
                        break  # 성공적으로 로딩되면 탈출
                    except:
                        print("🔄 상품 목록을 다시 불러오는 중...")
                        time.sleep(2)

            except Exception as e:
                print(f"❌ 오류 발생: {e}")
                continue  # 다음 상품으로 진행

        # 다음 페이지로 이동 (JavaScript 실행)
        try:
            page += 1  # 다음 페이지
            driver.execute_script(f"cmPageMove('{page}')")  # JavaScript 실행
            time.sleep(3)  # 페이지 로딩 대기
        except:
            print("❌ 더 이상 다음 페이지가 없습니다.")
            break  # 페이지가 없으면 종료

except Exception as e:
    print("❌ 크롤링 중 오류 발생:", e)

# 브라우저 종료
driver.quit()
```

```python
import os
import csv
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re

# CSV 파일 설정
csv_filename = "reviews_full.csv"
csv_columns = ["상품 URL", "상품 이름", "이미지 URL", "리뷰 ID", "별점", "리뷰 내용"]

# CSV 파일 초기화
if not os.path.exists(csv_filename):
    with open(csv_filename, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(csv_columns)

# 크롬 드라이버 실행
driver = webdriver.Chrome()

# "전체상품" 페이지 열기
base_url = 'https://www.mychef.kr/product/list?depth1=All'
driver.get(base_url)
time.sleep(3)

page = 1
clicked_urls = set()

try:
    while True:
        print(f"\n🚀 {page} 페이지 크롤링 시작")

        product_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
        )
        product_urls = [p.get_attribute("href") for p in product_elements]

        for index, product_url in enumerate(product_urls):
            if product_url in clicked_urls:
                continue

            try:
                driver.get(product_url)
                time.sleep(2)

                # 상품 이름
                try:
                    product_name = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, ".goods-tit"))
                    ).text.strip()
                except:
                    product_name = "상품 이름 없음"

                # 상품 이미지 URL
                try:
                    image_element = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, "img.lozad"))
                    )
                    image_name = image_element.get_attribute("src")

                    if image_url:
                        image_name = os.path.join(image_folder, f"{index + 1}_image.jpg")
                        img_data = requests.get(image_name).content
                        with open(image_name, "wb") as handler:
                            handler.write(img_data)
                        print(f"🖼️ 이미지 저장 완료: {image_name}")
                except Exception as e:
                    print(f"❌ 이미지 저장 실패: {e}")
                # try:
                #     image_element = driver.find_element(By.CSS_SELECTOR, "img.lozad")
                #     image_url = image_element.get_attribute("src")
                # except:
                #     image_url = "이미지 없음"

                # print(f"🛒 {product_name} / 📸 이미지: {image_url}")

                # 리뷰 요소 찾기
                review_elements = driver.find_elements(By.CSS_SELECTOR, "li.review-item")
                reviews = []

                for review_element in review_elements:
                    # 리뷰 ID 앞자리
                    try:
                        user_name = review_element.find_element(By.CSS_SELECTOR, ".name.user-name").text.strip()
                        review_id = user_name[:5]
                    except:
                        review_id = "ID 없음"

                    # 리뷰 별점
                    try:
                        star = review_element.find_element(By.CSS_SELECTOR, ".pointReviewContent")
                        style = star.get_attribute("style")
                        match = re.search(r'width:\s*(\d+(?:\.\d+)?)px', style)
                        star_rating = round(float(match.group(1)) / 20, 1) if match else "별점 없음"
                    except:
                        star_rating = "별점 없음"

                    # 리뷰 내용
                    try:
                        review_text = review_element.find_element(By.CSS_SELECTOR, "p.review.text span").text.strip()
                    except:
                        review_text = "내용 없음"

                    reviews.append([product_name, image_name, review_id, star_rating, review_text])

                # CSV 저장
                if reviews:
                    with open(csv_filename, "a", newline="", encoding="utf-8-sig") as f:
                        writer = csv.writer(f)
                        writer.writerows(reviews)
                    print(f"💬 리뷰 {len(reviews)}개 저장 완료")
                else:
                    print("ℹ️ 리뷰 없음")

                clicked_urls.add(product_url)
                driver.back()
                time.sleep(2)

            except Exception as e:
                print(f"❌ 상품 크롤링 오류: {e}")
                continue

        # 다음 페이지로 이동
        try:
            page += 1
            driver.execute_script(f"cmPageMove('{page}')")
            time.sleep(3)
        except:
            print("✅ 마지막 페이지 도달")
            break

except Exception as e:
    print(f"❌ 전체 크롤링 중 오류: {e}")

driver.quit()
```

```python
# 이미지 파일명 덮어쓰기 방지
import os
import csv
import requests
import re
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 이미지 저장 폴더 설정
thumbnail_folder = "product_images"
detail_folder = "product_detail_images"
os.makedirs(thumbnail_folder, exist_ok=True)
os.makedirs(detail_folder, exist_ok=True)

# CSV 설정
csv_filename = "reviews_including_empty.csv"
csv_columns = ["상품 이름", "이미지 파일명", "ID", "별점", "리뷰 내용"]
if not os.path.exists(csv_filename):
    with open(csv_filename, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(csv_columns)

# Selenium 드라이버 실행
driver = webdriver.Chrome()
driver.get("https://www.mychef.kr/product/list?depth1=All")
time.sleep(3)

page = 1
clicked_urls = set()
global_index = 1  # 이미지 파일명 중복 방지를 위한 전역 인덱스

try:
    while True:
        print(f"\n🚀 {page} 페이지 크롤링 시작")

        product_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[href^='/product/view?productCd=']"))
        )
        product_urls = list(set([p.get_attribute("href") for p in product_elements]))

        for product_url in product_urls:
            if product_url in clicked_urls:
                continue

            try:
                driver.get(product_url)
                time.sleep(2)

                # 상품 이름
                try:
                    product_name = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, ".goods-tit"))
                    ).text.strip()
                except:
                    product_name = "상품 이름 없음"

                # 썸네일 이미지 저장
                try:
                    img = driver.find_element(By.CSS_SELECTOR, "img.lozad")
                    thumb_url = img.get_attribute("src")
                    thumb_name = os.path.join(thumbnail_folder, f"{global_index}_image.jpg")
                    if thumb_url.startswith("http"):
                        with open(thumb_name, "wb") as f:
                            f.write(requests.get(thumb_url).content)
                        print(f"🖼️ 썸네일 저장 완료: {thumb_name}")
                    else:
                        thumb_name = "이미지없음"
                except:
                    print("❌ 썸네일 저장 실패")
                    thumb_name = "이미지없음"

                # 상세 이미지 (첫 번째) 저장
                try:
                    detail_img = driver.find_element(By.CSS_SELECTOR, "div.inner-content.productCont img")
                    detail_url = detail_img.get_attribute("src")
                    detail_name = os.path.join(detail_folder, f"{global_index}_detail.jpg")
                    if detail_url.startswith("http"):
                        with open(detail_name, "wb") as f:
                            f.write(requests.get(detail_url).content)
                        print(f"📘 상세 이미지 저장 완료: {detail_name}")
                except:
                    print("⚠️ 상세 이미지 없음")

                # 리뷰 수집
                review_page = 1
                seen_ids = set()
                all_reviews = []

                while True:
                    time.sleep(2)
                    reviews = driver.find_elements(By.CSS_SELECTOR, "li.review-item")
                    new = []

                    for r in reviews:
                        rid = r.get_attribute("data-id")
                        if rid in seen_ids:
                            continue
                        seen_ids.add(rid)

                        try:
                            user = r.find_element(By.CSS_SELECTOR, ".name.user-name").text.strip()
                            uid = user[:5]
                        except:
                            uid = ""

                        try:
                            star = r.find_element(By.CSS_SELECTOR, ".pointReviewContent").get_attribute("style")
                            match = re.search(r'width:\s*(\d+(?:\.\d+)?)px', star)
                            rating = round(float(match.group(1)) / 19, 1) if match else "-"
                        except:
                            rating = ""

                        try:
                            text = r.find_element(By.CSS_SELECTOR, "p.review.text span").text.strip()
                        except:
                            text = ""

                        new.append([product_name, f"{global_index}_image.jpg", uid, rating, text])

                    if new:
                        all_reviews.extend(new)
                        print(f"💬 리뷰 {len(new)}개 수집됨")

                    # 다음 리뷰 페이지
                    try:
                        review_page += 1
                        driver.execute_script(f"refreshReviewPaging('{review_page}')")
                        WebDriverWait(driver, 5).until(
                            lambda d: any(
                                r.get_attribute("data-id") not in seen_ids
                                for r in d.find_elements(By.CSS_SELECTOR, "li.review-item")
                            )
                        )
                    except:
                        print("📄 리뷰 끝")
                        break

                # 리뷰 없을 경우 기본 저장
                if not all_reviews:
                    all_reviews.append([product_name, f"{global_index}_image.jpg", "", "", ""])

                # CSV 저장
                with open(csv_filename, "a", newline="", encoding="utf-8-sig") as f:
                    writer = csv.writer(f)
                    writer.writerows(all_reviews)

                print(f"✅ 저장 완료: {len(all_reviews)}개")
                clicked_urls.add(product_url)
                global_index += 1  # 다음 상품 인덱스 증가
                driver.back()
                time.sleep(2)

            except Exception as e:
                print(f"❌ 상품 오류: {e}")
                continue

        # 다음 상품 페이지로 이동
        try:
            page += 1
            driver.execute_script(f"cmPageMove('{page}')")
            time.sleep(3)
        except:
            print("✅ 모든 상품 페이지 완료")
            break

except Exception as e:
    print(f"❌ 전체 오류 발생: {e}")

driver.quit()
```

```python
from surprise import SVD, NMF, KNNBasic, KNNWithMeans, BaselineOnly, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

# ----------------------------------------
# 1. 데이터 불러오기 및 전처리
# ----------------------------------------
# 데이터프레임 준비 (이미 df가 있다고 가정)
# 필요한 컬럼만 선택
df = df[['리뷰 ID', '상품 이름', '별점']]

# 별점 NaN 제거 + 숫자로 변환
df = df.dropna(subset=['별점'])
df['별점'] = pd.to_numeric(df['별점'], errors='coerce')
df = df.dropna(subset=['별점'])

# ----------------------------------------
# 2. Surprise Dataset 생성
# ----------------------------------------
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['리뷰 ID', '상품 이름', '별점']], reader)

# 학습/테스트 데이터 분할
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# ----------------------------------------
# 3. 다양한 알고리즘 학습 및 RMSE 비교
# ----------------------------------------
algorithms = {
    "SVD": SVD(),
    "NMF": NMF(),
    "KNNBasic": KNNBasic(),
    "KNNWithMeans": KNNWithMeans(),
    "BaselineOnly": BaselineOnly()
}

results = {}

print("📊 알고리즘별 RMSE 비교\n-------------------------")
for name, algo in algorithms.items():
    model = algo
    model.fit(trainset)
    predictions = model.test(testset)
    error = rmse(predictions, verbose=False)
    results[name] = error
    print(f"{name} RMSE: {round(error, 4)}")

# ----------------------------------------
# 4. 가장 성능 좋은 모델 선택
# ----------------------------------------
best_model_name = min(results, key=results.get)
print(f"\n✅ 가장 성능 좋은 모델: {best_model_name}")

# 선택된 모델로 다시 학습
best_model = algorithms[best_model_name]
best_model.fit(trainset)

# ----------------------------------------
# 5. 추천 함수 정의
# ----------------------------------------
def get_recommendations(user_id, top_n=4):
    user_items = df[df['리뷰 ID'] == user_id]['상품 이름'].unique()
    all_items = df['상품 이름'].unique()
    unseen_items = [item for item in all_items if item not in user_items]

    pred_ratings = [(item, best_model.predict(user_id, item).est) for item in unseen_items]
    top_recommendations = sorted(pred_ratings, key=lambda x: x[1], reverse=True)[:top_n]
    return top_recommendations

# ----------------------------------------
# 6. 추천 실행 예시
# ----------------------------------------
user_id = 'cjh4'  # 여기에 실제 유저 ID 입력
recommendations = get_recommendations(user_id)

print(f"\n🎯 사용자 '{user_id}' 추천 결과:")
for item, score in recommendations:
    print(f"- {item}: 예상 별점 {round(score, 2)}")
```

```python
import pandas as pd
from transformers import pipeline

# CSV 불러오기
df = pd.read_csv("mergedDf.csv", encoding="cp949")

# 컬럼명 자동 탐색
for col in df.columns:
    if '리뷰' in col and '내용' in col:
        review_col = col
        break

# 결측 제거 및 리뷰 상위 N개 추출
df = df.dropna(subset=[review_col])
review_texts = df[review_col].head().tolist()

# 감성 분석 모델 로딩
sentiment_pipeline = pipeline("sentiment-analysis", model="snunlp/KR-FinBert-SC")

# 감성 분석 수행
results = [sentiment_pipeline(text)[0] for text in review_texts]

# 길이 확인해서 안전하게 값 저장
df.loc[df.index[:len(results)], '감성 결과'] = [res['label'] for res in results]
df.loc[df.index[:len(results)], '감성 점수'] = [res['score'] for res in results]

# 결과 출력
print(df[[review_col, '감성 결과', '감성 점수']].head(len(results)))
```

```python
import pandas as pd
from surprise import SVD, NMF, KNNBasic, KNNWithMeans, BaselineOnly, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

# 1. CSV 불러오기
df = pd.read_csv("감성분석결과.csv", encoding="utf-8-sig")
df = df[['리뷰 ID', '상품 이름', '별점', '감성 결과', '감성 점수']].dropna()

# 2. 협업 필터링 데이터 준비
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['리뷰 ID', '상품 이름', '별점']], reader)
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# 3. 알고리즘 비교 및 선택
algorithms = {
    "SVD": SVD(),
    "NMF": NMF(),
    "KNNBasic": KNNBasic(),
    "KNNWithMeans": KNNWithMeans(),
    "BaselineOnly": BaselineOnly()
}
best_model = None
best_rmse = float('inf')

for name, algo in algorithms.items():
    model = algo
    model.fit(trainset)
    predictions = model.test(testset)
    error = rmse(predictions, verbose=False)
    if error < best_rmse:
        best_rmse = error
        best_model = model

# 4. 감성 점수 계산 (상품별 평균 감성 점수)
# 긍정: 1.0, 부정: 0.0으로 환산
df['감성 점수 정제'] = df['감성 결과'].map(lambda x: 1.0 if 'positive' in x.lower() else 0.0)
sentiment_scores = df.groupby('상품 이름')['감성 점수 정제'].mean().to_dict()

# 5. 하이브리드 추천 함수
def hybrid_recommend(user_id, top_n=5, alpha=0.7):
    user_items = df[df['리뷰 ID'] == user_id]['상품 이름'].unique()
    all_items = df['상품 이름'].unique()
    unseen_items = [item for item in all_items if item not in user_items]

    hybrid_scores = []

    for item in unseen_items:
        cf_score = best_model.predict(user_id, item).est
        senti_score = sentiment_scores.get(item, 0.5)  # 감성 점수 없으면 0.5로 간주
        final_score = alpha * cf_score + (1 - alpha) * senti_score * 5  # 감성 점수 0~1 → 0~5 환산
        hybrid_scores.append((item, final_score))

    top_items = sorted(hybrid_scores, key=lambda x: x[1], reverse=True)[:top_n]
    return top_items

# 6. 예시 추천
print("🎯 추천 결과:")
print(hybrid_recommend("kenn", top_n=5))
```
