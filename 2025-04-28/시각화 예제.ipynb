```python
# 기상 관측 이래 대전의 최고 기온이 높았던 날과 기온은
f = open('daejeon.csv')
data = csv.reader(f)
next(data)
max_temp = 0.0
max_date = ''
cnt = 0
result = []
for a in data:
    if cnt >= 1:
        if a[-1] == '':
            a[-1] = -999
        a[-1] = float(a[-1])
        if max_temp < a[-1]:
            max_temp = a[-1]
            max_date = a
    cnt += 1

print(max_temp)
print(max_date)
```

```python
# 8월에 해당하는 데이터만 추출하여 그래프를 그려보세요.
# 그래프 색깔은 'hotpink'

f = open('daejeon.csv')
data = csv.reader(f)
result = []
cnt = 0
for row in data:    
    if cnt >= 1:
        if row[-1] != '':
            if row[0].split('-')[1] == '08':
                result.append(row[-1])
    cnt += 1
f.close()

plt.figure(figsize =(12,3))
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 매년 돌아오는 각자의 생일에 최고기온 데이터를 표현해 보세요.
f = open('daejeon.csv')
data = csv.reader(f)
result = []
cnt = 0
for row in data:    
    if cnt >= 1:
        if row[-1] != '':
            if row[0].split('-')[1] == '04' and row[0].split('-')[2] == '10':
                result.append(row[-1])
    cnt += 1
f.close()

plt.figure(figsize =(12,5))
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 자신의 생일 년도 이후의 최고기온 데이터를 표현해 보세요.
f = open('daejeon.csv')
data = csv.reader(f)
result = []
cnt = 0
for row in data:    
    if cnt >= 1:
        if row[-1] != '':
            if row[0].replace('\t','').split('-')[0] >= '1993' and row[0].split('-')[1] == '04' and row[0].split('-')[2] == '10':
                result.append(float(row[-1]))
    cnt += 1
f.close()

plt.figure(figsize =(12,5))
plt.plot(result, 'hotpink')
plt.show()
```

```python
f = open('daejeon.csv')
data = csv.reader(f)
result = []
result2 = []
cnt = 0
for row in data:    
    if cnt >= 1:
        if row[-1] != '':
            if row[0].replace('\t','').split('-')[0] >= '1993' and row[0].split('-')[1] == '04' and row[0].split('-')[2] == '10':
                result.append(float(row[-1]))
                result2.append(float(row[-2]))
    cnt += 1
f.close()
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize =(12,5))
plt.plot(result, 'hotpink', label = '데이터1' )
plt.plot(result2, 'skyblue', label = '데이터2')
plt.title('기온')
plt.legend()  # 범례 표시
plt.show()
```

```python
# 8월 일별 기온 데이터를 상자 그림으로 표현하기
f = open('daejeon.csv')
data = csv.reader(f)
next(data)
cnt = 0

day = []
for i in range(31):
    day.append([])

for row in data:
    if cnt >= 1:
        if row[-1] != '':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
    cnt += 1
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.boxplot(day, showfliers = False)
plt.show()
```
