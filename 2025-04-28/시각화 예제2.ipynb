```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
               for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols

fig, ax = plt.subplots(figsize=(9, 8), dpi = 300)

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
   row = i % nrows
   col = i // nrows
   y = Y - (row * h) - h

   xi_line = w * (col + 0.05)
   xf_line = w * (col + 0.25)
   xi_text = w * (col + 0.3)

   ax.text(xi_text, y, name, fontsize=(10),
           horizontalalignment='left',
           verticalalignment='center')

   ax.hlines(y + h * 0.1, xi_line, xf_line,
             color=colors[name], linewidth=(6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                   top=1, bottom=0,
                   hspace=0, wspace=0)
plt.show()
```

# 유임 승차 비율이 가장 높은 역은 어디일까?
-rate = 유인승차인원/무인승차인원
1. 데이터를 읽어온다.
2. 모든 역의 데이터를 바탕으로 각 역의 비율(rate)을 계산한다.
3. 비율이 가장 높은 역을 찾는다. 그 비율이 얼마인지 출력한다

```python
f = open('subwayfee_d.csv')
data = csv.reader(f)
next(data)
rate = 0
mx = 0
max_station = ''
for row in data:
    if int(row[6]) != 0:
        rate = int(row[4]) / int(row[6])
        if rate > mx:
            mx = rate
            max_station = row[3]
            
print(mx)
print(max_station)
```

```python
import csv
import matplotlib.pyplot as plt

class SubwayDataAnalyzer:
    def __init__(self, filename):
        self.filename = filename
        self.data = []
    
    def load_data(self):
        with open(self.filename, 'r', encoding='utf-8') as f:
            reader = csv.reader(f)
            next(reader)
            self.data = [row for row in reader]
            for row in self.data:
                row[4:] = list(map(int, row[4:]))
    
    def find_max_station(self, start_index=8, end_index=10):
        max_passengers = 0
        max_station = ""
        
        for row in self.data:
            passengers = sum(row[start_index:end_index])
            if passengers > max_passengers:
                max_passengers = passengers
                max_station = row[2]
        
        return max_station, max_passengers


analyzer = SubwayDataAnalyzer('subwaytime_dj.csv')
analyzer.load_data()
station, max_passengers = analyzer.find_max_station()

print(f"가장 붐비는 역: {station}")
print(f"최대 승객 수: {max_passengers}")
```
