```python
import matplotlib.pyplot as plt

plt.plot([10,20,30,40])
plt.show()
```

```python
plt.title('TITLE')
plt.plot([1,2,3,4],[12,43,25,15])
```

```python
plt.plot([10,20,30,40])
plt.plot([40,30,20,10])
plt.show()
```

```python
plt.title('line Style')
plt.plot([10,20,30,40], color = 'red', linestyle='--', label='dashad')
plt.plot([40,30,20,10], color = 'black', ls=':', label='dotted')
plt.legend()
plt.show()
```

```python
#   
plt.title('Marker')
plt.plot([10,20,30,40], 'r.', label='circle')
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend()
plt.show()
```
