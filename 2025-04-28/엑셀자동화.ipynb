```python
import openpyxl as op
# ver.GPT
# 각 파일을 불러오는 경로
files = [
    r'C:\Users\admin\Desktop\엑셀자동화\거래명세서.xlsx',
    r'C:\Users\admin\Desktop\엑셀자동화\거래명세서2.xlsx',
    r'C:\Users\admin\Desktop\엑셀자동화\거래명세서3.xlsx'
]

# 통합 워크북 생성
wb_combined = op.Workbook()
ws_combined = wb_combined.active
ws_combined.title = "통합 거래명세서"

# 헤더 추가
headers = ["작성일자", "발행번호", "거래처", "제품명", "규격", "수량", "단가", "금액"]
ws_combined.append(headers)

# 병합 셀의 값을 안전하게 읽는 함수
def get_merged_cell_value(ws, cell):
    for merge_range in ws.merged_cells.ranges:
        # 병합 범위를 확인하고, 병합된 범위 안에 있는 셀인지 체크
        if cell.coordinate in merge_range:
            # 병합 범위의 첫 번째 셀을 기준으로 값을 가져옴
            min_col, min_row, max_col, max_row = merge_range.bounds
            return ws.cell(row=min_row, column=min_col).value
    # 병합되지 않은 셀의 경우 그냥 값 반환
    return cell.value

# 각 파일의 데이터를 읽어오기
for file in files:
    wb = op.load_workbook(file)
    ws = wb.active

    # 데이터 추출 (병합 셀 처리 포함)
    작성일자 = get_merged_cell_value(ws, ws["B10"])  # 작성일자
    발행번호 = get_merged_cell_value(ws, ws["C10"])  # 발행번호
    거래처 = get_merged_cell_value(ws, ws["B2"])     # 거래처

    # 제품명, 규격, 수량, 단가, 금액 반복 추출 (14행부터 시작)
    for row in ws.iter_rows(min_row=14, max_row=19, min_col=2, max_col=6, values_only=True):
        if row[0] is not None:  # 제품명이 있으면 데이터를 추가
            제품명, 규격, 수량, 단가, 금액 = row
            ws_combined.append([작성일자, 발행번호, 거래처, 제품명, 규격, 수량, 단가, 금액])

# 통합된 데이터를 저장
wb_combined.save(r'C:\Users\admin\Desktop\엑셀자동화\통합거래명세서.xlsx')

print("데이터 통합 완료!")
```

```python
# 파일 통합 코드.
import openpyxl as op
# from pathlib import Path
# folder = 'C:/Users/admin/Desktop/엑셀자동화'
# folder_path = Path(folder)
# execel_files = folder_path.glob('거래명세서*.xlsx')
# for f in execel_files:
#     print(f)

def excel_data_extractor(file):

    wb = op.load_workbook(file, data_only = True)
    ws = wb.active

    col1 = ws['A9'].value # 작성일자
    col2 = ws['C9'].value # 발행번호
    col3 = ws['A2'].value # 거래처
    col_list = [col1, col2, col3]
    col4 = ws['B13:F13'] # 제품명, 규격, 수량, 단가, 금액
    for col in col4:
        for cell in col:
            col_list.append(cell.value)


    date = ws['A10'].value # 작성일자 데이터
    date = date.strftime('%Y-%m-%d') # YYYY-mm-dd 포맷으로 수정
    issue_num = ws['C10'].value # 발행번호 데이터
    company = ws['B2'].value # 거래처 데이터

    rows = [] # 한 행씩 완성시킬 리스트

    for i in range(6):
        data_row = [date, issue_num, company] # 한 행 데이터
        flag = False

        for k in range(5):
            contents = ws.cell(row= 14 + i, column=2 + k).value # 세부내용 출력
            if contents != None: # 결측치 처리
                flag = True # 세부내용 있는 경우만 행 추가
                data_row.append(contents)

        if flag == True:
            rows.append(data_row) # 한 행씩 추가

    return col_list, rows

file1 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서.xlsx'
file2 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서2.xlsx'
file3 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서3.xlsx'
files = [file1, file2, file3]

new_wb = op.Workbook()
new_ws = new_wb.active

header = False
for file in files:
    col_list, rows = excel_data_extractor(file)

    if header == False:
        new_ws.append(col_list)
        header = True

    for row in rows:
        new_ws.append(row)
new_wb.save(r'C:\Users\admin\Desktop\엑셀자동화\통합거래명세서3.xlsx')
```

```python
import pandas as pd

def excel_data_extractor(file):

    df = pd.read_excel(file)

    col1 = df.iloc[7,0] # 작성일자
    col2 = df.iloc[7,2] # 발행번호
    col3 = df.iloc[0,0] # 거래처
    col_list = [col1, col2, col3]
    col_content = list(df.iloc[11, 1:6]) #제품명,규격,수량,단가,금액
    col_list += col_content

    date = df.iloc[8,0]
    date = date.strftime('%Y-%m-%d')
    issue_num = df.iloc[8,2]
    company = df.iloc[0,1]

    df_new = pd.DataFrame(columns = col_list)
    df_new[col_list[0]] = [date] * 6
    df_new[col_list[1]] = [issue_num] * 6
    df_new[col_list[2]] = [company] * 6

    for k in range(5):
        df_new[col_list[3+k]] = df.iloc[12:12+6, 1+k].values

    df_new = df_new.dropna()

    return df_new

file1 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서.xlsx'
file2 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서2.xlsx'
file3 = r'C:\Users\admin\Desktop\엑셀자동화\거래명세서3.xlsx'
files = [file1, file2, file3]

new_excel = pd.DataFrame()
for file in files:
    df = excel_data_extractor(file)
    new_excel = pd.concat([new_excel , df], ignore_index = True)

new_excel.to_excel(r'C:\Users\admin\Desktop\엑셀자동화\거래명세서_통합_판다스1.xlsx',index= False)
```
