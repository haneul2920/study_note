```python
import csv
import matplotlib.pyplot as plt

class SubwayTimeAnalyzer:
    def __init__(self, filename):
        self.filename = filename
        self.data = []
        self.s_in = [0] * 24
        self.s_out = [0] * 24

    def load_data(self):
        with open(self.filename, 'r', encoding='cp949') as f:
            reader = csv.reader(f)
            next(reader)
            self.data = [row for row in reader]
            for row in self.data:
                row[4:] = list(map(int, row[4:]))

    def analyze_data(self):
        for row in self.data:
            for j in range(24):
                if row[3] == '승차':
                    self.s_in[j] += row[j + 4]
                elif row[3] == '하차':
                    self.s_out[j] += row[j + 4]

    def plot_data(self):
        plt.rc('font', family='Malgun Gothic')
        plt.figure(dpi=100)
        plt.title('지하철 시간대별 승하차 인원 추이')
        plt.plot(self.s_in, label='승차')
        plt.plot(self.s_out, label='하차')
        plt.legend()
        plt.xticks(range(24), range(4, 28))
        plt.show()


analyzer = SubwayTimeAnalyzer('subwaytime_dj.csv')
analyzer.load_data()
analyzer.analyze_data()
analyzer.plot_data()
```

```python
import csv
import matplotlib.pyplot as plt

class SubwayDataAnalyzer:
    def __init__(self, filename):
        self.filename = filename
        self.data = []
    
    def load_data(self):
        with open(self.filename, 'r', encoding='utf-8') as f:
            reader = csv.reader(f)
            next(reader)
            self.data = [row for row in reader]
            for row in self.data:
                row[4:] = list(map(int, row[4:]))
    
    def find_max_station(self, start_index=8, end_index=10):
        max_passengers = 0
        max_station = ""
        
        for row in self.data:
            passengers = sum(row[start_index:end_index])
            if passengers > max_passengers:
                max_passengers = passengers
                max_station = row[2]
        
        return max_station, max_passengers


analyzer = SubwayDataAnalyzer('subwaytime_dj.csv')
analyzer.load_data()
station, max_passengers = analyzer.find_max_station()

print(f"가장 붐비는 역: {station}")
print(f"최대 승객 수: {max_passengers}")
```
